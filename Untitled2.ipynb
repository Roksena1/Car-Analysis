**Visualizations:**

```python
sns.histplot(df_car['city mpg'],bins=30)
plt.title('Distribution for the City Mpg')
plt.xlabel('City MPG')
plt.ylabel('Frequency')
plt.show()
```

```python
avg_msrp_per_size = df_car.groupby('Vehicle Size', as_index=False)['MSRP'].mean()
plt.figure(figsize=(5,5))
sns.barplot(x ='Vehicle Size',
            y= 'MSRP', data=avg_msrp_per_size, hue='Vehicle Size',palette='coolwarm')
plt.title('Average MSRP by Vehicle Size')
plt.xlabel('Vehicle Size')
plt.ylabel('Average MSRP')
plt.show()
```

```python
sns.scatterplot(x='Engine HP',
            y='MSRP', data=df_car)
plt.title('Relationship between Engine HP and MSRP')
plt.xlabel('Engine HP')
plt.ylabel('MSRP')
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

```python
sns.boxplot(x='Driven_Wheels', y='MSRP', data=df_car, hue='Driven_Wheels', palette='coolwarm')
plt.yscale('log')
plt.title('Distribution of MSRP for each category Driven_Wheels')
plt.xlabel('Driven_Wheels')
plt.ylabel('MSRP')
plt.show()
```

```python
df_filtered = df_car[df_car['highway MPG'] <= 120]
avg_mpg = df_filtered.groupby('Transmission Type', as_index=False)[['city mpg', 'highway MPG']].mean()
avg_mpg = avg_mpg.sort_values('Transmission Type')
plt.figure(figsize=(5,6))
plt.plot(avg_mpg['Transmission Type'], avg_mpg['highway MPG'],
         marker='s', color='red', label='Highway MPG')
plt.plot(avg_mpg['Transmission Type'], avg_mpg['city mpg'],
         marker='o', color='blue', label='City MPG')
plt.title('Average City MPG and Highway MPG for different Transmission Type')
plt.xlabel('Transmisson Type')
plt.ylabel('Average MPG')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

**Correlation Analysis:**

```python
correlation = ['Engine HP', 'MSRP', 'Popularity', 'city mpg', 'highway MPG']
df_corr = df_car[correlation]
corr_matrix = df_corr.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Selected Variables", fontsize=14)
plt.show()
```

**Barplot**

```python
plt.figure(figsize=(41,6))
sns.barplot(x='Make', y='Engine HP', data=df_car, hue='Make', palette='coolwarm')
plt.title('Distribution of HP across the car brand',fontsize=17)
plt.xlabel('Make',fontsize=16)
plt.ylabel('Engine HP',fontsize=16)
plt.show()
```

Dies bedeutet, dass die meisten Autos im normalen Marktsegment (Alltagsautos) liegen. Es gibt einige Autos mit extrem hohem MSRP (ber 1.000.000 USD), aber nicht unbedingt die hchste HP haben.
