**Visualizations:**

```python
sns.histplot(df_car['city mpg'],bins=30)
plt.title('Distribution for the City Mpg')
plt.xlabel('City MPG')
plt.ylabel('Frequency')
plt.show()
```

```python
avg_msrp_per_size = df_car.groupby('Vehicle Size', as_index=False)['MSRP'].mean()
plt.figure(figsize=(5,5))
sns.barplot(x ='Vehicle Size',
            y= 'MSRP', data=avg_msrp_per_size, hue='Vehicle Size',palette='coolwarm')
plt.title('Average MSRP by Vehicle Size')
plt.xlabel('Vehicle Size')
plt.ylabel('Average MSRP')
plt.show()
```

```python
sns.scatterplot(x='Engine HP',
            y='MSRP', data=df_car)
plt.title('Relationship between Engine HP and MSRP')
plt.xlabel('Engine HP')
plt.ylabel('MSRP')
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

```python
sns.boxplot(x='Driven_Wheels', y='MSRP', data=df_car, hue='Driven_Wheels', palette='coolwarm')
plt.yscale('log')
plt.title('Distribution of MSRP for each category Driven_Wheels')
plt.xlabel('Driven_Wheels')
plt.ylabel('MSRP')
plt.show()
```

```python
df_filtered = df_car[df_car['highway MPG'] <= 120]
avg_mpg = df_filtered.groupby('Transmission Type', as_index=False)[['city mpg', 'highway MPG']].mean()
avg_mpg = avg_mpg.sort_values('Transmission Type')
plt.figure(figsize=(5,6))
plt.plot(avg_mpg['Transmission Type'], avg_mpg['highway MPG'],
         marker='s', color='red', label='Highway MPG')
plt.plot(avg_mpg['Transmission Type'], avg_mpg['city mpg'],
         marker='o', color='blue', label='City MPG')
plt.title('Average City MPG and Highway MPG for different Transmission Type')
plt.xlabel('Transmisson Type')
plt.ylabel('Average MPG')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

**Correlation Analysis:**

```python
correlation = ['Engine HP', 'MSRP', 'Popularity', 'city mpg', 'highway MPG']
df_corr = df_car[correlation]
corr_matrix = df_corr.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Selected Variables", fontsize=14)
plt.show()
```

**Zusammenfassung**

Analysierend die Daten über die Autos ist es in der „Gruppierung Analyse“ zu sehen, dass Large-Fahrzeuge den höchsten durchschnittlichen MSRP (der vom Hersteller empfohlene Preis des Autos) haben (53621 USD), gefolgt von Midsize-Fahrzeuge (39056 USD) und anschließend die Compact-Fahrzeuge mit 34327 USD, was es auf ein logisches Muster hindeutet, da die Large-Fahrzeuge mehr Rohstoffe benötigen (z. B. Stahl, Aluminium, Kunststoff), was zu höheren Produktionskosten führt. Large-Fahrzeuge haben oft größere und leistungsstärkere Motoren, die teuer in der Produktion sind. Large-Fahrzeuge sind oft mit mehr Sicherheitsfunktionen ausgestattet, da sie für Langstrecken oder den Familiengebrauch gedacht sind, was eine höhere Zahlungsbereitschaft bedeutet. Die Popularität zeigt, dass die Large-Fahrzeuge am beliebtesten sind, während Midsize-Fahrzeuge am wenigsten populär sind. Das Histogramm „Distribution for the City Mpg“ zeigt,  dass die meisten Fahrzeige eine City MPG zwischen ca. 15 und 25. Fahrzeuge mit höheren City MPG-Werten über 40 kommen selten vor. Es gibt einige wenige extreme Werte über 100 MPG. Die Verteilung ist rechtsschief (positive Skewness), was bedeutet, dass es mehr Fahrzeuge mit niedrigerer City MPG gibt und nur wenige mit sehr hoher City MPG.
Der Scatter Plot zeigt die Beziehung zwischen der Motorleistung (Engine HP, x-Achse) und dem MSRP(y-Achse). Es ist eine positive Korrelation zu sehen: Autos mit höherer Motorleistung haben höheren MSRP. Die meisten Fahrzeuge haben weniger als 400 HP und kosten unter 100.000 USD. Dies bedeutet, dass die meisten Autos im normalen Marktsegment (Alltagsautos) liegen. Es gibt einige Autos mit extrem hohem MSRP (über 1.000.000 USD), aber nicht unbedingt die höchste HP haben. Diese Punkte könnten Luxus- oder Supersportwagen sein. Die Boxplot-Grafik zeigt die Verteilung des MSRP (empfohlener Verkaufspreis) für Fahrzeuge mit unterschiedlichen Antriebssystemen. "Rear Wheel Drive" (Hinterradantrieb) zeigt eine breite Verteilung der Preise mit einer höheren Medianlinie, was darauf hindeutet, dass viele Premium- oder Luxusfahrzeuge diesen Antrieb verwenden. "Front Wheel Drive" (Vorderradantrieb) hat den niedrigsten Medianpreis, was darauf hindeutet, dass viele günstigere Fahrzeuge diese Antriebsart nutzen. "All Wheel Drive" (Allradantrieb) hat auch höheren Median, was darauf hindeutet, dass es auch in dieser Kategorie teurere Wagen enthalten sind. Die Preise sind auf einer logarithmischen Skala dargestellt. Dies wurde gewählt, weil MSRP-Werte große Unterschiede aufweisen, von wenigen Tausend bis über eine Million USD. Bei "Rear Wheel Drive" und "All Wheel Drive" gibt es zahlreiche extreme Ausreißer über 100.000 USD. Dies könnten Sportwagen und Luxuslimousinen sein. Die anderen Kategorien haben auch Ausreißer, aber nicht in demselben Umfang. Das Liniendiagramm zeigt die Durchschnittswerte von City MPG und Highway MPG für verschiedene Getriebetypen. "Direct Drive" hat die höchsten Werte für City und Highway MPG, was darauf hinweist, dass diese Fahrzeuge besonders effizient sind. Automatisierte und manuelle Getriebe haben niedrigere MPG-Werte als Direct Drive. Der allgemeine Trend ist dass, die rote Linie (Highway MPG) über der blauen Linie (City MPG) liegt, da Fahrzeuge auf der Autobahn in der Regel effizienter sind. Nur Direct Drive ist effizienter in der Stadt als auf dem Autobahn. Elektrofahrzeuge haben oft ein Direct Drive System. Die Korrelationsmatrix zeigt, dass MSRP positiv mit Engine HP korreliert (+0.66) ist. Das ergibt Sinn, da teurere Fahrzeuge oft leistungsstärkere Motoren haben. Negative Korrelation zwischen Engine HP und MPG (-0.47 für City, -0.41 für Highway),weil stärkere Motoren (mehr PS) haben tendenziell einen höheren Kraftstoffverbrauch, was sich in einer negativen Korrelation äußert. Popularität hat kaum Einfluss auf andere Variablen (0,04). Das bedeutet, dass beliebte Autos nicht unbedingt sparsamer oder leistungsstärker sind. Die letzte Grafik zeigt die Verteilung von HP auf die Automarke. Bugatti, McLaren, Maybach und Lamborghini haben die höchsten HP, was es auch logisch ist, weil sie Sport- und Luxuswagen sind.
